```python
#import library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import matplotlib
```

```python
df1=pd.read_csv('/content/drive/MyDrive/DatasetDA/Retail/customer_shopping_data.csv')
df1
```

```python
df1.info()
```

```python
df1.isnull().sum()
```

```python
df1['invoice_date']=pd.to_datetime(df1['invoice_date'])
df1.info()
```

```python
df1['month']=df1['invoice_date'].dt.month
df1['year']=df1['invoice_date'].dt.year
```

```python
df1.loc[df1['age']<=25, 'age_group'] = '<25'
df1.loc[df1['age'].between(26,35), 'age_group'] = '26-35'
df1.loc[df1['age'].between(36,45), 'age_group'] = '36-45'
df1.loc[df1['age'].between(46,55), 'age_group'] = '46-55'
df1.loc[df1['age'].between(56,65), 'age_group'] = '56-65'
df1.loc[df1['age']>65, 'age_group'] = '>65'
```

# Gender

```python
df1['revenue']=df1['price']*df1['quantity']
df1
```

```python
df_gender=pd.DataFrame({'count':df1.groupby('gender')['customer_id'].count(),
                        'revenue':df1.groupby('gender')['revenue'].sum()
}).reset_index()
df_gender
```

```python
fig, ax =plt.subplots(ncols=2, figsize=(15,7))
ax[0].pie(df_gender['revenue'],labels=df_gender['gender'],autopct='%.1f%%')
ax[0].set_title('Revenue by gender')

ax[1]=sns.barplot(data=df_gender, x='gender', y='count')
ax[1].set_title("Number of customer by gender")
```

It shows that the percentage and the number of females is higher than men. We can say that women take care of their family so they usually go shopping than men.

```python
plt.figure(figsize=(10, 5))
ax = sns.countplot(data=df1, x='category', hue='gender')
plt.title('Categories bought by gender')
```

It shows that the number of females is higher than the proportion of men in all category. It is also worth pointing out that clothing is the category that most bought by customers.

```python
new_shop = df1.groupby('gender')['shopping_mall'].value_counts().to_frame(name='count').reset_index()
new_shop
```

```python
plt.figure(figsize=(15,4))
ax = sns.barplot(data=new_shop, x='shopping_mall', y='count', hue='gender')
plt.title("Count of clients visit shopping by gender")
```

It shows that in all city in this dataset, the number of female customer is higher than men. In addition, Kanyon and Mall of Istanbul are city has the highest the number of customer.

# Category

```python
df_cate=df1.groupby('category')['quantity'].sum().reset_index()
df_cate
```

```python
g = sns.catplot(data=df1, x='category', kind='count', order=df1.category.value_counts().index)
plt.xticks(rotation=45, ha="right")
g.fig.suptitle("Category buy of clients")
g.set_xlabels("Categories")
g.set_ylabels("Quantity")
plt.subplots_adjust(top=0.9)
```

As I said previously, clothing is the most bought category. Then we have cosmetic and food and beverage. Books is the category which has the lowest sold quantity.

```python
plt.pie(df_cate['quantity'],labels=df_cate['category'],autopct='%.1f%%')
plt.title('% Quantity by category')
```

# Time

```python
df_year_month_revenue=df1.groupby(['year','month'])['revenue'].sum().reset_index()
df_year_month_revenue
```

```python
sns.relplot(kind='line', data=df_year_month_revenue, x='month', y='revenue', hue='year', aspect=2.5, marker='o', estimator='sum')
```

The above chart shows that the trend of revenue in 2021 and 2020 is similar and they do not fluctuate significantly. And because now is April so the revenue after April 2023 in this data set may be predicted revenue or and mistake.

# Malls

```python
df_mall=pd.DataFrame({'clients':df1.groupby('shopping_mall')['customer_id'].count(),
                      'revenue':df1.groupby('shopping_mall')['revenue'].sum()
    
}).reset_index()
df_mall
```

```python
plt.pie(df_mall['revenue'],labels=df_mall['shopping_mall'],autopct='%.1f%%')
plt.title('% Revenue by mall')
```

```python
ax = sns.barplot(data=df_mall.sort_values(by='clients',ascending=False), x='clients',y='shopping_mall')
plt.title("Quantity customers by shopping mall")
plt.xlabel("Number of customers")
plt.show()
```

Both above chart show that the Mall of Istanbul and Kanyon makes the most revenue and has the highest number of customer. We can conclue that these 2 cities is big and modern. So when the business should prioritize the market in these city by diversify and provide more products there.

# Payment

```python
df_payment=df1.groupby('payment_method')['invoice_no'].count().reset_index()
df_payment
```

```python
plt.pie(df_payment['invoice_no'],labels=df_payment['payment_method'],autopct='%.1f%%')
plt.title('%Payment method by orders')
```

```python
plt.figure(figsize=(10, 5))
ax = sns.countplot(data=df1, x='payment_method', hue='gender')
plt.title('Payment method paid by gender')
```

As we can see, cash is the most popular payment method, followed by credit card and debit card.
